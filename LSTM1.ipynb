```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import keras
from __future__ import print_function
from keras.models import
```

```python
df = pd.read_csv('train_pre.csv')
```

```python
df_1=df.iloc[:,1:12]
df_norm = (df_1 - df_1.mean()) / (df_1.max() - df_1.min())
```

```python
df_norm.columns
```

```python
values=df_norm.values
values
```

```python
X= values[:,0]
X=X.reshape(len(X),1,1)
y = df['Choice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=60)
n_batch=1
model = Sequential()
model.add(LSTM(10, activation='tanh',batch_input_shape=(n_batch,X.shape[1],X.shape[2]),stateful=True))
model.add(Dense(1))
print(model.summary())
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100, batch_size=n_batch, verbose=1, shuffle= False)
# Final evaluation of the model
scores = model.evaluate(X_test, y_test, batch_size=n_batch,verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```
